These steps will create a minimal-sized model based on the distilbert-base-uncased architecture. You can modify the architecture to suit your needs, but keep in mind that smaller architectures will have fewer capabilities and may not perform as well as larger models.

```python
from transformers import pipeline

model = pipeline('question-answering', model='distilbert-base-cased-distilled-squad', tokenizer='distilbert-base-cased-distilled-squad')
result = model(context='The Mona Lisa is a 16th century oil painting created by Leonardo da Vinci. It is held at the Louvre in Paris.', question='Wheres is the Louvre?')
print(result['answer'])
```

```python
from transformers import pipeline

model_name = "distilbert-base-uncased-distilled-squad"
qa_pipeline = pipeline("question-answering", model=model_name)

question = "Which are major cities of Australia?"
context = "Australia is a country and continent surrounded by the Indian and Pacific oceans. Its major cities – Sydney, Brisbane, Melbourne, Perth, Adelaide – are coastal. Its capital, Canberra, is inland. The country is known for its landscape of mountains, forests, beaches, and reefs."

result = qa_pipeline(question=question, context=context)

print(f"Answer: {result['answer']}")
print(f"Score: {result['score']}")
print(f"Start index: {result['start']}")
print(f"End index: {result['end']}")

```
